#Artificial Neural Network

##Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

##Part 1 - Data Preprocessing

###Importing the dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 8 - Deep Learning/Section 39 - Artificial Neural Networks (ANN)/Python/Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

###Encoding categorical data

Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X[:, 2] = label_encoder.fit_transform(X[:, 2])
```

One Hot Encoding the "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

###Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

###Feature Scaling

```python
#Feature scaling is applied to all the variables when we develop ANN irrespective of whether it is encoded or it is categorical variable
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

##Part 2 - Building the ANN

###Initializing the ANN

```python
'''Ann is created as an object of the Sequential class which allows ann to be developed as a sequence of layers as opposed to computational graph.
Computational graph example is Boltzman machines. '''

ann = tf.keras.models.Sequential()
```

###Adding the input layer and the first hidden layer

```python
#Any hidden fully connected layer is added by using Dense class. 
#Dropout layer avoids overfitting
#units correspond to number of hidden neurons in the layer we are gonna add
#input layer has all the features as input neurons. One of them as each neuron.
ann.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
```

###Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

###Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
#when there are more than 2 categories, activation function must not be sigmoid instead it should be softmax
```

##Part 3 - Training the ANN

###Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

#best optimizer is that can perform gradient descent. It will update weights in order to reduce loss error
#whenever we perform binary classification, loss function should always be binary cross entropy 
#for non binary classification, the loss function must be categorical cross entropy. 
```

###Training the ANN on the Training set

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
#number of elements used for comparing the prediction with the actual value is specified by batch size
```

##Part 4 - Making the predictions and evaluating the model

###Predicting the result of a single observation

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

###Predicting the Test set results

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

###Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
