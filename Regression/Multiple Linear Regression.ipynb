# Multiple Linear Regression

## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 5 - Multiple Linear Regression/Python/50_Startups.csv')
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
column_Transformer = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder = 'passthrough')
```

```python
x = np.array(column_Transformer.fit_transform(x))
```

```python
print(x)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Predicting the Test set results

```python
y_test_predicted = regressor.predict(x_test)
```

```python
y_test_predicted = np.array(y_test_predicted)
y_test = np.array(y_test)
whole = np.stack([y_test_predicted, y_test], axis=-1)
np.set_printoptions(precision = 2)
pd.DataFrame(data = whole, columns = ('y_test_predicted', 'y_test'))
```

```python
np.set_printoptions(precision = 4)
np.set_printoptions(suppress = True)
print(regressor.coef_)
print(regressor.intercept_)
```
