#Decision Tree Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 8 - Decision Tree Regression/Python/Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Decision Tree Regression model on the whole dataset


```python
from sklearn.tree import DecisionTreeRegressor
decisionTree_Regressor = DecisionTreeRegressor(random_state = 1)
decisionTree_Regressor.fit(x, y)
```

## Predicting a new result

```python
decisionTree_Regressor.predict([[6.5]])
```

## Visualising the Decision Tree Regression results (higher resolution)

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'blue')
plt.plot(x_grid, decisionTree_Regressor.predict(x_grid), color = 'black')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
