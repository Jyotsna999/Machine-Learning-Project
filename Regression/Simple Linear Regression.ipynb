# Simple Linear Regression

# Importing the Libraries



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the Dataset


```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv')
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

# Splitting the dataset into Training set & Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

# Training the SLR model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train) 
```

# Predicting the Training set results

```python
y_train_predicted = regressor.predict(x_train)
```

# Predicting the Test set results


```python
y_test_predicted = regressor.predict(x_test)
```

# Visualising Training set results

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, y_train_predicted, color = 'black')
plt.title("Salary Vs Experience in Training set")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualising Test set results

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, y_train_predicted, color = 'black')
plt.title("Salary Vs Experience in Test set")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
